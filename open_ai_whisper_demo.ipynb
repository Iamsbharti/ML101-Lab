```python
import whisper
import whisper
from whisper.utils import get_writer
from tabulate import tabulate

# Load the model
# Get the available models
available_models = whisper.available_models()

# Prepare data for tabulate
table_data = [(idx + 1, model) for idx, model in enumerate(available_models)]

# Print the table
print(tabulate(table_data, headers=["#", "Model Name"], tablefmt="pretty"))
```

```python
# Load the model
model = whisper.load_model("large")
```

```python
# Define the audio file path
audio_path = "First Flight.mp3"
```

```python
# Transcribe the audio
import time
start_time = time.time()

result = model.transcribe(audio_path)
output_directory = "./output"

end_time = time.time()
execution_time = end_time - start_time

print("Transcription Execution time:", execution_time, "seconds")
```

ran-usage.PNG

```python
# Save as an SRT file
srt_writer = get_writer("srt", output_directory)
srt_writer(result, audio_path)
print('SRT Transcript Created')
```

```python
# Save as a VTT file
vtt_writer = get_writer("vtt", output_directory)
vtt_writer(result, audio_path)
print('VTT Transcript  Created')
```

```python
# Save as a JSON file
json_writer = get_writer("json", output_directory)
json_writer(result, audio_path)
print('JSON Transcript Created')
```
